#### Function to filter places based on 'gps'

```python
def places_filter(dataset, lat_l, lat_h, lon_l, lon_h):
    reviews_us = []
    for d in dataset:
        if (d['gps'] != None and d['price'] != None):
            if ((d['gps'][0] >= lat_l) \
              & (d['gps'][0] <= lat_h) \
             & (d['gps'][1] >= lon_l) \
            & (d['gps'][1] <= lon_h)):
                reviews_us.append(d)
    return reviews_us 
```
